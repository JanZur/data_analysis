- Poisson distribution

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

```

```python
#create a Gaussian distribution
mu = 0
sigma = 1
x = np.linspace(-5, 5, 100)
y = stats.norm.pdf(x, mu, sigma)
plt.plot(x, y)
plt.show()
```

```python
# binomial distribution
# used for experiments that have two possible outcomes, success or failure
# e.g. coin toss, pass/fail, win/lose
# when number of trials is known

n = 10
p = 0.5
x = np.arange(0, 11)
y = stats.binom.pmf(x, n, p)
plt.plot(x, y)
plt.show()
```

```python
#create a Poisson distribution
# poisson distribution is used for experiments that are independent and have a constant rate of success
# e.g. number of phone calls received in a day or radioactive decay
# when number of failures is not known


mu = 1.5
measured = 6
x = np.arange(0, 10)
y = stats.poisson.pmf(x, mu, loc=2)
plt.plot(x, y)
plt.show()
```
